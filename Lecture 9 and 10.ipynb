# List Traversal Function

```python
def get_report_df(report_list):
    
         # Save an empty mega DF
        mega_report_df = pd.DataFrame()
         # For loop
         # For each Class Dictionary in the list:
        for division in report_list:
              # Make a scores DF
            scores_df = pd.DataFrame(division["scores"])
                # Make an Emails DF
            emails_df = pd.DataFrame(division["emails"])
            emails_df["student_email"] = emails_df["student_email"].str.lower()
                # Merge the two DF
            scores_and_emails_df = pd.merge(scores_df, emails_df, how = "left", on = ["student_name"])    
                # Add class 
            scores_and_emails_df["class"] = division["class"]
                # Add division 
            scores_and_emails_df["division"] = division["division"]
                # Add teacher
            scores_and_emails_df["teacher"] = division["teacher"]
                # Add it to the mega DF
            mega_report_df = pd.concat([mega_report_df, scores_and_emails_df], ignore_index = True)
        # Total and average scores    
        mega_report_df["total_scores"] = mega_report_df["math_scores"] + mega_report_df["information_technology_scores"] + mega_report_df["physics_scores"] + mega_report_df["chemistry_scores"]
        mega_report_df["average_scores"] = mega_report_df["total_scores"]/4
        # Status column
        mega_report_df.loc[(mega_report_df["average_scores"] < 40), "status"] = "Fail"
        mega_report_df.loc[(mega_report_df["average_scores"] >= 40), "status"] = "Pass"
        # Character grade
        mega_report_df.loc[(mega_report_df["average_scores"]>=90), "character_grade"] = "A"
        mega_report_df.loc[(mega_report_df["average_scores"]>=70) & (mega_report_df["average_scores"]<90), "character_grade"] = "B"
        mega_report_df.loc[(mega_report_df["average_scores"]>=50) & (mega_report_df["average_scores"]<70), "character_grade"] = "C"   
        mega_report_df.loc[(mega_report_df["average_scores"]>=40) & (mega_report_df["average_scores"]<50), "character_grade"] = "D"
        mega_report_df.loc[(mega_report_df["average_scores"]< 40), "character_grade"] = "F"       
    # Return the mega DF    
        return mega_report_df
```

```python
main_report = get_report_df(report_database)
main_report
```

## How many students passed and failed in every division

```python
pass_fail_df = main_report.groupby(["status", "division"]).agg(total_students = ("status", "count"))
pass_fail_df
```

## Mean total score by division

```python
mean_total_score_df = main_report.groupby(["division"]).agg(total_score = ("total_scores", "mean"))
mean_total_score_df
```

## Lowest physics score

```python
lowest_grade = main_report.groupby(["division"]).agg(lowest_physics_grade = ("physics_scores", "min"))
lowest_grade
```

## Highest score

```python
highest_grade = main_report.groupby(["division"]).agg(highest_physics_grade = ("physics_scores", "max"))
highest_grade
```

```python
pcm_average_score = main_report.groupby(["division"]).agg(avg_math_score = ("math_scores","mean"),
                                                          avg_chemistry_score = ("chemistry_scores", "mean"),
                                                          avg_physics = ("physics_scores","mean"))
pcm_average_score
```
